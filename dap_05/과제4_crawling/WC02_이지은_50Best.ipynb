Main Page

```python
def crawl_main_page():
    
    Ranking   = []
    Cafe      = []
    Menu      = []
    Link      = []
    
    url = 'https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
    html = urlopen(url)
    soup = BeautifulSoup(html, "lxml")
    
    con_tag = soup.find('div', 'content post')
    sammy_tags = con_tag.find_all('div', 'sammy')
    
    for sammy_tag in sammy_tags[0:49]:
        
        rank_tag = sammy_tag.find('div', 'sammyRank')
        rank = int(rank_tag.get_text())
        rank
        
        cafe_tag = sammy_tag.find('b')
        cafe = cafe_tag.get_text()
        cafe
        
        menu_tag = sammy_tag.find('a').contents[2]
        menu = menu_tag.strip()
        menu
        
        link = 'https://www.chicagomag.com/' + sammy_tag.find('a')['href']
        link
        
        Ranking.append(rank)
        Cafe.append(cafe)
        Menu.append(menu)
        Link.append(link)
        
    data = {'순위':Ranking, '카페명':Cafe, '메뉴명':Menu, '링크':Link}
    ret_df = pd.DataFrame(data)
    
    return ret_df
```

```python
df = crawl_main_page()
df
```

```python
df.set_index('순위', inplace = True)
df
```

```python
df.to_csv('data/TheBestSandwiches.csv', sep=',', encoding='UTF-8')
```
