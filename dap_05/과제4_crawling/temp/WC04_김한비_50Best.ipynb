## 데이터 프레임안에 정보 넣기

```python
def crawl_chicago1_rank():
    
    Ranking   = []
    Menu     = []
    Cafe = []
    Link = []
    
    url = 'http://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
    html = urlopen(url)
    soup = BeautifulSoup(html, "lxml")

    top_Rank = soup.find_all('div', 'sammyRank')
    top_sites = soup.find_all('div', 'sammyListing')

    for top_Rank in top_Rank[0:50]:
        ranking = int(top_Rank.get_text())
        Ranking.append(ranking)
        
    for top_sites in top_sites[0:50]:   
        menu = top_sites.find('b').get_text()
        Menu.append(menu)
        
        cafe = top_sites.get_text().split("\n")[1]
        Cafe.append(cafe)
        
        link =  top_sites.find('a')['href']
        
        # 몇몇 링크는 http 가 붙은 상태로 저장되있음. 
        if link[0] == 'h':
            link =  top_sites.find('a')['href']
        else:
            link = 'http://www.chicagomag.com' + top_sites.find('a')['href']
        Link.append(link)
        
        
    data1   = {'순위':Ranking, '메뉴': Menu, '카페': Cafe, '링크': Link}
    ret_df = pd.DataFrame(data1)
    
    return ret_df  
```

```python
df1 = crawl_chicago1_rank()
df1
```

```python
# 홈페이지 주소가 없는경우 nonetype 뜨기때문에 try-catch로 null값 지정. 

try:
    homepage = soup1.find_all('em')[0].find('a')['href']
except:
    homepage = "null"
homepage
```

```python
def crawl_chicago2_rank():
    
    Price   = []
    Phone = []
    Homepage = []
    Address = []

    for link in range(0,50):
        url1 = df1['링크'][link]
        html = urlopen(url1)
        soup1 = BeautifulSoup(html, "lxml")
        
        # 12번의 경우 em 태그가 5개라 결과가 다르게 나옴 
        if len(soup1.find_all('em')) == 4:
            site1 = soup1.find_all('em')[0]
        else:
            site1 = soup1.find_all('em')[1]

        price = site1.get_text().split(",")[0].split('.')
       
        #센트가 있는것 없는것 나누어 분류 
        if len(price[1]) == 2:
            price = price[0] + '.' + price[1]
        else :
            price = price[0] + '.' + '00'
        
        #
        if str(soup1.find_all('em')[0]).find("Multiple locations")>0:
            address = "Multiple Location"
        else: 
            price1 = site1.get_text().split(",")[0].split('.')
            if len(price1[1]) == 2:
                address = price1[2:]           
            else :
                address = price1[1:]
            address = ''.join(address)
            
        # 전화번호는 항상 - 가지기 때문에, 없는건 Multiple Location으로 지정
        if str(site1).find("-") > 0:
            pos = str(site1).find("-")
            phone = str(site1)[pos-3:pos+8]
        else:
            phone = 'Multiple Location'
            
        # 홈페이지가 없는것도 있어서, 없을땐 null로 지정     
        try:
            homepage = soup1.find_all('em')[0].find('a')['href']
        except:
            homepage = "null"
       
    
        Price.append(price)
        Phone.append(phone)
        Address.append(address)
        Homepage.append(homepage)
        
    data2   = {'가격':Price, '전화번호':Phone,'주소':Address,'홈페이지주소':Homepage}
    ret_df = pd.DataFrame(data2)
    
    return ret_df  
```

```python
df2 = crawl_chicago2_rank()
df2
```
