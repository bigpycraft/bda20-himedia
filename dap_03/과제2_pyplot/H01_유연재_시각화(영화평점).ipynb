```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib import font_manager, rc

# font_name = font_manager.FontProperties(fname="/Library/Fonts/AppleGothic.ttf").get_name()
font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/MALGUN.TTF").get_name()
rc('font', family=font_name) #한글 이미지 때문에 설정해준다, 작은 글씨가 가장 깔끔하게 나온다

plt.rcParams['figure.figsize'] = (10,6)

%matplotlib inline
```

```python
# 바챠트
t = np.arange(10)
x = [8.78, 8.34, 8.53, 5.0, 8.52, 8.5, 9.67, 9.0, 9.04, 9.21] #관람객
y = [5.71, 5.0, 5.0, 6.0, 6.67, 5.0, 4.0, 5.83, 7.0, 6.0] #기자&평론가
z = [7.12, 6.08, 7.51, 6.92, 8.18, 4.61, 9.58, 6.94, 8.39, 9.3] #네티즌

plt.figure(figsize=(20,8))
plt.bar(t, x, color='deeppink', width=0.2, label='관람객')
plt.bar(t + 0.25, y, color='greenyellow', width=0.2,label='기자&평론가')
plt.bar(t + 0.5, z, color='royalblue', width=0.2,label='네티즌')
plt.title('네이버 영화 평점')
plt.xlabel('영화 제목')
plt.ylabel('영화 평점')
plt.legend()
plt.grid()

plt.xticks(t+0.25, ("조제", "이웃사촌", "도굴", "미드나이트 스카이", "런", "800", "스웨그", "더 프롬", "삼진그룹영어토익반", "파티마의 기적"))   # 값 전체를 수정보다는 눈금 값을 알맞게 맞춰준다
plt.show()
```

```python
# 파이차트 ver1

label = "스웨그", "파티마의 기적", "삼진그룹영어토익반", "런", "도굴", "조제", "더 프롬", "미드나이트 스카이", "이웃사촌", "800"
score = [9.58, 9.3, 8.39, 8.18, 7.51, 7.12, 6.94, 6.92, 6.08, 4.61]

fig1, ax1 = plt.subplots(figsize=(7,7))
ax1.pie(score, labels=label, startangle=90, counterclock=False)

plt.title('네티즌 영화 평점 비율')
plt.show()
```

```python
# 파이차트 ver2
label = "스웨그", "파티마의 기적", "삼진그룹영어토익반", "런", "도굴", "조제", "더 프롬", "미드나이트 스카이", "이웃사촌", "800"
score = [9.58, 9.3, 8.39, 8.18, 7.51, 7.12, 6.94, 6.92, 6.08, 4.61]
explode = (0.15, 0, 0, 0, 0, 0, 0, 0, 0, 0)

fig1, ax1 = plt.subplots(figsize=(7,7))
ax1.pie(score, explode=explode, labels=label, autopct='%1.1f%%', shadow=True, startangle=90, counterclock=False)
ax1.axis('equal')  

plt.title('네티즌 영화 평점 비율')
plt.show()
```
